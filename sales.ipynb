```python
import pandas as pd  
import numpy as np  
import seaborn as sns  
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv(r"C:\\Users\\NET\\Desktop\\data work\\data\\sales.csv", encoding='latin1')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns= ['YEAR_ID','MONTH_ID','ADDRESSLINE1','ADDRESSLINE2','STATE','TERRITORY'], inplace=True)
```

```python
data
```

```python
data['CONTACTFULLNAME'] = data['CONTACTFIRSTNAME'].astype(str)+' '+data['CONTACTLASTNAME'].astype(str)
data
```

```python
data.drop(columns=['CONTACTFIRSTNAME'], inplace=True)
```

```python
data.drop(columns=['CONTACTLASTNAME'], inplace=True)
```

```python
data.head(2)
```

```python
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'])
```

```python
data.head(2)
```

```python
data['SALES'] = data['QUANTITYORDERED'] * data['PRICEEACH']
```

```python
data['SALES'] = data['SALES'].round(0)
data
```

```python
ymgroup = data.groupby(pd.Grouper(key='ORDERDATE', freq='m'))['SALES'].sum().reset_index()
ymgroup
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=ymgroup,x='ORDERDATE', y='SALES')
plt.xticks(rotation= 90)
plt.show()
```

```python
statprd = data.pivot_table(index=pd.Grouper(key='ORDERDATE', freq='m'),columns='STATUS', values='QUANTITYORDERED', aggfunc=sum)
statprd
```

```python

fig, ax = plt.subplots()
fig.patch.set_visible(True)
ax.axis('off')
table = ax.table(cellText=statprd.values, colLabels=statprd.columns, loc='center',rowLabels=statprd.index)
plt.show()
```
